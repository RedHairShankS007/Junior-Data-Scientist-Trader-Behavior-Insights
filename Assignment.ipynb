```python
import pandas as pd
# Load datasets
sentiment_df = pd.read_csv('fear_greed_index.csv')  # Columns: Date, Classification
trader_df = pd.read_csv('historical_data.csv')    # Columns include: time, account, symbol, execution price, etc.
```

```python
sentiment_df.head(5)
```

```python
# Step 1: Convert 'Date' in sentiment_df to datetime format
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'])
```

```python
sentiment_df.head(5)
```

```python
sentiment_df.info()
```

```python
trader_df.head(5)
```

```python
#step-3
# Convert to datetime
trader_df["Timestamp IST"] = pd.to_datetime(trader_df["Timestamp IST"], format="%d-%m-%Y %H:%M")

# Extract only the date part
trader_df["Date"] = trader_df["Timestamp IST"].dt.date

# If you want just 'YYYY-MM-DD' string format instead of datetime.date
trader_df["Date_str"] = trader_df["Timestamp IST"].dt.strftime("%Y-%m-%d")
```

```python
trader_df.head()
```

```python
trader_df.info()
```

```python
trader_df.rename(columns={"Date": "date"}, inplace=True)
```

```python
trader_df['date'] = pd.to_datetime(trader_df['date'])
```

```python
trader_df.info()
```

```python
trader_df.drop("Date_str", axis=1, inplace=True)
```

```python
# Step 4: Merge datasets on date
merge_df = pd.merge(trader_df, sentiment_df, left_on='date', right_on='date', how='left')
```

```python
from sklearn.preprocessing import MinMaxScaler

# Select columns to normalize
cols_to_normalize = ['Closed PnL', 'Size Tokens', 'Size USD', 'Fee', 'Start Position', 'value']

# Initialize scaler
scaler = MinMaxScaler()

# Apply normalization
normalized_data = scaler.fit_transform(merge_df[cols_to_normalize])

# Add normalized columns to the DataFrame
for i, col in enumerate(cols_to_normalize):
    merge_df[col + '_norm'] = normalized_data[:, i]
```

```python
merge_df.info()
```

```python
merge_df.head()
```

```python
# Standardize side values
merge_df['Side'] = merge_df['Side'].str.lower()
```

```python
# Create a new column for trade direction
merge_df['position_type'] = merge_df['Side'].map({
    'buy': 'Long',
    'sell': 'Short'
})
```
